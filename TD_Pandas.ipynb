# Tutorial 1: What is Pandas?

Pandas est l'une des bibliothèques les plus importantes de Python.

Pandas dispose de structures de données pour faciliter l'analyse des données. Les plus utilisées sont les structures de données Series et DataFrame. La structure de données Series est unidimensionnelle, c'est-à-dire qu'elle est constituée d'une colonne. La structure de données DataFrame est bidimensionnelle, c'est-à-dire qu'elle est constituée de lignes et de colonnes.

Pour installer Pandas, vous pouvez utiliser "pip install pandas".


```python
import pandas as pd # Importons pandas avec pd
```

```python
f_eq[f_eq<=0.6]=0.5 # Attribuer la valeur de 0.5 aux valeurs <= 0.6
f_eq
```

```python
"Moto" in f_eq # Si "Moto" est en f_eq --> True
```

```python
"VL" in f_eq # Si "VL" est en f_eq --> True
```

```python
f_eq/5 # diviser par 5
```

```python
f_eq**2 # f_eq au carré
```

```python
f_eq.isnull() # S'il y a des valeurs nulles --> True
```

# Tutoriel 4 : DataFrame

Qu'est-ce qu'un DataFrame ?

```python
import pandas as pd
```

```python
data={"veh_ID":["veh.1","veh.2","veh.3","veh.4"],      
      "Vitesse":[40,30,10,50],      
      "Typologie":["Moto","Bus","Trottinette","VL"],      
      "Deux Roues":["Y","N","Y","N"]}
```

```python
df=pd.DataFrame(data) # définir dataframe
df
```

```python
df=pd.DataFrame(data,columns=["veh_ID","Vitesse","Typologie","Deux Roues","Longueur"])# Définir etiquettes  
df
```

```python
df=pd.DataFrame(data,columns=["Vitesse","Typologie","Deux Roues","Longueur"],
               index=["veh.1","veh.2","veh.3","veh.4"]) # si pas de données NaN
df
```

```python
df["Typologie"] # afficher une colonne specifique
```

```python
my_columns=["Typologie","Vitesse"]
df[my_columns]# afficher deux colonnes specifiques
```

```python
df.Vitesse #afficher une colonne specifique(une autre façon)
```

```python
df.loc[["veh.1"]] # afficher une ligne
```

```python
df.loc[["veh.1","veh.4"]] # afficher deux lignes
```

```python
df["Longeur"]=5 #affecter 5 à toutes les lignes de la colonne
df
```

```python
values=[2,12,1.5,4.5]
df["Longeur"]=values #affecter un vector à une colonne
df
```

```python
df["vitesse_elevee"]=df.Vitesse>=35 #if Vitesse>=35 --> True
df
```

# Tutoriel 5 : Indexation-Sélection-Filtrage dans la bibliothèque Pandas

```python
data={"veh_ID":["veh.1","veh.2","veh.3","veh.4"],      
      "Vitesse":[40,30,10,50],      
      "Typologie":["Moto","Bus","Trottinette","VL"],      
      "Deux Roues":["Y","N","Y","N"]}
```

```python
df["Typologie"]
```

```python
df[["Typologie","Vitesse"]]
```

```python
df[:2] #obtenir des données jusqu'à la deuxième ligne
```
