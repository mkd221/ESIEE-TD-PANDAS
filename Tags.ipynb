```python
import pandas as pd
```

# Tags


```python
file = "esiee v1__09-18-2023_11h_40m_06s.csv"
```

```python
cols = ['Timestamp', 'EPC', 'TID', 'Antenna', 'RSSI', 'Frequency', 'Hostname', 'PhaseAngle', 'DopplerFrequency']
```

```python
tags = pd.read_csv(file, skiprows=3, sep=";", names = cols)
tags
```

```python

tags = tags[['Timestamp', 'EPC', 'Antenna', 'RSSI']]
tags
```
